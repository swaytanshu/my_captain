```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('D:\Study Materials\selected\Artificial Intellingence\image_recognition\mnist_train.csv')
test = pd.read_csv('D:\Study Materials\selected\Artificial Intellingence\image_recognition\mnist_test.csv')
```

```python
# reshaping the extracted data into qa reasonable size
a = a.reshape(28,28).astype('uint8')
plt.imshow(a)
```

```python
# preparing training data
X_train = dataset.iloc[:, 1:].values
y_train = dataset.iloc[:, 0].values
```

```python
# preparing testing data
X_test = test.iloc[:, 1:].values
y_test = test.iloc[:, 0].values
```

```python
# Fitting random forest classifier to the Training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 300, criterion = 'entropy')
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
# visualizing the predicted results 
pd.DataFrame(y_pred)
```

```python
# printing the correct, incorrect predictions and accuracy of the model 
count = 0
for i in range(len(y_pred)):
    if y_pred[i] == y_test[i]:
        count = count + 1
print(f"correct predictions are  : {count} out of {len(y_pred)}")
print(f"incorrect predictions are : {len(y_pred) - count} ")
print(f"so the accuracy is {(count / len(y_pred)) * 100}")
```
